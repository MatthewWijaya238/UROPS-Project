```python
import pandas as pd
```

```python
train_df = pd.read_csv(r"D:\MATTHEW\UROPS\Datasets\First Batch\First Batch Experiment 2.csv", index_col = 0)
test_df = pd.read_csv(r"D:\MATTHEW\UROPS\Datasets\Second Batch\Second Batch Experiment 2.csv", index_col = 0)
```

```python
test_df.columns
```

```python
for target_col in train_df.columns:
    print(target_col)
    other_cols = [col for col in train_df.columns if col != target_col]    
    X_train = train_df[other_cols]
    y_train = train_df[target_col]
    X_test = test_df[other_cols]
    y_test = test_df[target_col]

    model = LinearRegression()
    sfs = SequentialFeatureSelector(model)
    sfs.fit(X_train, y_train)
    X_train_new = sfs.transform(X_train)
    X_test_new = sfs.transform(X_train)
    print(sfs.get_support())
```

```python
X_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, -1].values
X_test = test_df.iloc[:, :-1].values
y_test = test_df.iloc[:, -1].values
```

```python
model = LinearRegression()
sfs = SequentialFeatureSelector(model, cv=10, scoring= 'explained_variance', n_features_to_select=2)
```

```python
sfs.fit(X_train, y_train)
```

```python
sfs.get_support()
```

```python
X_train1 = sfs.transform(X_train)
X_test1 = sfs.transform(X_test)
```

```python
model.fit(X_train1, y_train)
```

```python
y1_pred = model.predict(X_test1)
print(y_test)
print(y1_pred)
```

```python
print("Intercept:", model.intercept_)  # b in y = mx + b
print("Coefficient:", model.coef_)    # m in y = mx + b
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y1_pred))
print("R² Score:", r2_score(y_test, y1_pred))
```
