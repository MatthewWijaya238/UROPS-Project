```python
def arff_to_csv(input_dir, output_dir):
    for file_name in os.listdir(input_dir):
        if file_name.endswith('.arff'):
            input_path = os.path.join(input_dir, file_name)
            output_path = os.path.join(output_dir, file_name[:-5])
            # print(input_path)
            # print(output_path)
            try:
                arff_file = arff.loadarff(input_path)
                df = pandas.DataFrame(arff_file[0])
                for col in df.select_dtypes([object]).columns:
                    df[col] = df[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
                    df.to_csv(output_path + '.csv', index=False)
            except Exception as e:
                print(f"Error processing {file_name}: {e}")
```

```python
def clean_arff_header(input_path):
    """Fixes invalid characters in the ARFF header and shortens the @relation name if needed."""
    with open(input_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    with open(input_path, 'w', encoding='utf-8') as f:
        for line in lines:
            if line.strip().startswith("@relation"):
                line = "@relation dataset\n"  # Replace with a generic name
            f.write(line)
```

```python
def remove_extra_columns(input_path, output_path):
    """Removes unexpected columns from an ARFF file."""
    with open(input_path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    # Count @attribute lines
    attribute_count = sum(1 for line in lines if line.lower().startswith("@attribute"))

    # Find the @data section
    data_index = next(i for i, line in enumerate(lines) if line.lower().startswith("@data")) + 1

    # Extract data rows
    data_rows = [line.strip().split(",") for line in lines[data_index:] if line.strip()]

    # Check if data has extra columns
    for i, row in enumerate(data_rows):
        if len(row) > attribute_count:  # Too many columns!
            print(f"⚠️ Row {i+1} has {len(row)} columns but should have {attribute_count}. Removing extra column.")
            data_rows[i] = row[-attribute_count:]  # Keep only the last `attribute_count` columns

    # Reconstruct the ARFF file
    with open(output_path, "w", encoding="utf-8") as f:
        f.writelines(lines[:data_index])  # Keep headers
        for row in data_rows:
            f.write(",".join(row) + "\n")
```
