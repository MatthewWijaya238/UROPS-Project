```python

def format_dataset(input_path, output_path, class_column = 'class'):
    df = pd.read_csv(input_path)
    # print(df)
    print(input_path)
    potential_class_columns = [col for col in df.columns if col.lower() in ['class', 'target']]
    class_column_name = potential_class_columns[0] if potential_class_columns else df.columns[-1]
    
    df.rename(columns={class_column_name: class_column}, inplace=True)
    class_col_data = df.pop(class_column)  # Remove class column
    df[class_column] = class_col_data      # Append it to the last column
    id_columns = [col for col in df.columns if 'id' == col.lower()]
    if id_columns:
        print(f"Removing ID column(s): {id_columns}")
        df = df.drop(columns=id_columns)
    empty_columns = [col for col in df.columns if 'Unnamed' in col or df[col].isna().all()]
    if empty_columns:
        print(f"Removing empty/unnecessary column(s): {empty_columns}")
        # Drop these columns
        df = df.drop(columns=empty_columns)
    df.replace('?', np.nan, inplace=True)
    
    # Convert columns to numeric (non-convertible values will become NaN)
    # df = df.apply(pd.to_numeric, errors='coerce')
    df.dropna(inplace=True)
    # print(df)
    binary_columns = [col for col in df.columns if df[col].nunique() == 2]
    df = convert_to_binary(df, binary_columns)
    # print(df)
    # if len(df.columns[:-1]) > 10:
    #     selected_features = random.sample(list(df.columns[:-1]), 10)
    #     selected_features.append('class')
    #     df = df[selected_features]
    df.to_csv(output_path, index=False)

def format_multiple_datasets(input_dir, output_dir, class_column_name='class'):
    """
    Preprocess multiple datasets in a directory.

    Args:
        input_dir (str): Directory containing input CSV files.
        output_dir (str): Directory to save preprocessed CSV files.
        class_column_name (str): Desired name for the class column.
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Loop through all CSV files in the input directory
    for file_name in os.listdir(input_dir):
        if file_name.endswith('.csv'):
            input_path = os.path.join(input_dir, file_name)
            output_path = os.path.join(output_dir, file_name)

            try:
                format_dataset(input_path, output_path, class_column_name)
            except Exception as e:
                print(f"Error processing {file_name}: {e}")
```

```python
def convert_to_binary(df, binary_columns):
    """
    Convert categorical binary columns into 0 and 1 format.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        binary_columns (list): List of column names to convert.

    Returns:
        pd.DataFrame: DataFrame with converted binary columns.
    """
    for col in binary_columns:
        unique_values = df[col].dropna().unique()
        if len(unique_values) != 2:
            raise ValueError(f"Column '{col}' is not binary (unique values: {unique_values})")
        
        # Map the unique values to 0 and 1
        df[col] = df[col].map({unique_values[0]: 0, unique_values[1]: 1})
        
        # Ensure no unmapped values exist
        if df[col].isnull().any():
            raise ValueError(f"Unmapped values found in column '{col}' after conversion.")
    
    return df
```

```python
def format_workbook(input_path, output_path, class_column = 'class'):
    df = pd.read_excel(input_path)
    # print(df)
    print(input_path)
    df.rename(columns={df.columns[-1]: class_column}, inplace=True)
    id_columns = [col for col in df.columns if 'id' == col.lower()]
    if id_columns:
        print(f"Removing ID column(s): {id_columns}")
        df = df.drop(columns=id_columns)
    empty_columns = [col for col in df.columns if 'Unnamed' in col or df[col].isna().all()]
    if empty_columns:
        print(f"Removing empty/unnecessary column(s): {empty_columns}")
        # Drop these columns
        df = df.drop(columns=empty_columns)
    df.replace('?', np.nan, inplace=True)
    
    # Convert columns to numeric (non-convertible values will become NaN)
    # df = df.apply(pd.to_numeric, errors='coerce')
    df.dropna(inplace=True)
    # print(df)
    binary_columns = [col for col in df.columns if df[col].nunique() == 2]
    df = convert_to_binary(df, binary_columns)
    # print(df)
    if len(df.columns[:-1]) > 10:
        selected_features = random.sample(list(df.columns[:-1]), 10)
        selected_features.append('class')
        df = df[selected_features]
    df.to_csv(output_path, index=False)
```

```python
def convert_multiclass_to_binary_balanced(input_dir, output_dir):
    for file_name in os.listdir(input_dir):
        input_path = os.path.join(input_dir, file_name)
        output_path = os.path.join(output_dir, file_name)
        try:
            df = pd.read_csv(input_path)
            class_counts = df['class'].value_counts(normalize=True).sort_values()
            classes = class_counts.index.tolist()
            proportions = class_counts.values

            # Step 2: Split classes into two groups (greedy approach)
            group_0, group_1 = [], []
            sum_0, sum_1 = 0, 0

            for i, cls in enumerate(classes):
                if sum_0 < sum_1:
                    group_0.append(cls)
                    sum_0 += proportions[i]
                else:
                    group_1.append(cls)
                    sum_1 += proportions[i]

            print(f"Group 0 classes: {group_0}, Total Proportion: {sum_0:.2f}")
            print(f"Group 1 classes: {group_1}, Total Proportion: {sum_1:.2f}")

            # Step 3: Apply the encoding
            df['binary_label'] = df['class'].apply(lambda x: 0 if x in group_0 else 1)

            # Step 4: Balance the dataset (downsample the larger group)
            df_0 = df[df['binary_label'] == 0]
            df_1 = df[df['binary_label'] == 1]

            min_size = min(len(df_0), len(df_1))

            df_0_balanced = resample(df_0, replace=False, n_samples=min_size, random_state=42)
            df_1_balanced = resample(df_1, replace=False, n_samples=min_size, random_state=42)

            df_balanced = pd.concat([df_0_balanced, df_1_balanced]).sample(frac=1, random_state=42).reset_index(drop=True)
            df_balanced = df_balanced.drop(columns=['class'])
            # print(df_balanced)
            df_balanced.rename(columns={'binary_label':'class'}, inplace=True)
            df_balanced.to_csv(output_path, index=False)
        except Exception as e:
                print(f"Error processing {file_name}: {e}")
```

```python
def balance_classes(input_dir, output_dir):
    for file_name in os.listdir(input_dir):
        input_path = os.path.join(input_dir, file_name)
        output_path = os.path.join(output_dir, file_name)
        try:
            df = pd.read_csv(input_path)
            X = df.drop(columns=['class'])  # Features
            y = df['class']  # Labels

            # Print initial class distribution
            print("Initial class distribution:")
            print(y.value_counts())

            # Define undersampler
            rus = RandomUnderSampler(random_state=42)
            X_resampled, y_resampled = rus.fit_resample(X, y)

            # Print new class distribution
            print("\nClass distribution after downsampling:")
            print(pd.Series(y_resampled).value_counts())

            # Create a new DataFrame
            df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
            df_resampled['class'] = y_resampled
            df_resampled.to_csv(output_path, index=False)
        
        except Exception as e:
                print(f"Error processing {file_name}: {e}")
```
